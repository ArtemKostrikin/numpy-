```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
t = np.linspace(0, 10, 51)
print(t)
```

```python
f = np.cos(t)
print(f)
```

```python
plt.plot(t, f, color = 'green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
print(x)
```

```python
y1 = x**2
print(y1)
```

```python
y2 = 2 * x + 0.5
print(y2)
```

```python
y3 = -3 * x - 1.5
print(y3)
```

```python
y4 = np.sin(x)
print(y4)
```

```python
fig, ax = plt.subplots(nrows=2, ncols =2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x. y4)
ax1.set_title('График $y_1$')
ax2.set_title('График $y_2$')
ax3.set_title('График $y_3$')
ax4.set_title('График $y_4$')
ax1.set_xlim([-5, 5])
ax1.set_size_inches(8, 6)
plt.subplots_adjust(wpace=0.3, hspace=0.3)
plt.show()
```
